```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = (240-30) / 2
sigma = (mu * 1.341) - mu 
arrival_times = np.random.normal(mu, sigma, 1000)
```

```python
count, bins, ignored = plt.hist(arrival_times, 30, density=True) 
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r') 
plt.show()
```
